```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
```

```python
train_path = 'DATASET/DATASET/TRAIN'
test_path = 'DATASET/DATASET/TEST'
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Activation
from tensorflow.keras.preprocessing.image import ImageDataGenerator,img_to_array, load_img
from tensorflow.keras.utils import plot_model
from glob import glob
```

```python

x_data = []
y_data = []

# Assuming categories is a list of folders (categories) that contain images
for category in glob(train_path+"/*"):
    for file in tqdm(glob(category+"/*")):  # Modify this pattern as per your file types
        img_array = cv2.imread(file)
        img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)  # Corrected color conversion
        x_data.append(img_array)
        y_data.append(category.split('/')[-1])  # Assuming category names are the labels

data = pd.DataFrame({'image': x_data, 'label': y_data})
```

```python
colors =["#a0d157", "#c48bb8" ]
plt.pie(data.label.value_counts(),labels=["organic","recyclable"],colors=colors,autopct='%0.2f%%',startangle=90,explode=(0.05,0.05),shadow=True)
plt.show()
```
